# Feature Engineering

```python
def day_density(x):
    
    density_num = [362, 363, 364, 365]
    
    if x in density_num:
        
        return 1
    
    else:
        return 0
```

```python
def get_year_day(x):
    
    if (x.is_leap_year) & (x >= dt.datetime(x.year, 4, 29)):
        
        return x.dayofyear - 1
    else:
        return x.dayofyear
```

```python
def get_features(df_temp):
    df_temp['date'] = pd.to_datetime(df_temp['date'])
    df_temp['day of week'] = df_temp['date'].dt.day_name()
    df_temp['month'] = df_temp['date'].dt.month
    df_temp['year'] = df_temp['date'].dt.year
    df_temp['day of year'] = df_temp['date'].apply(lambda x: get_year_day(x))
    df_temp['holiday density'] = df_temp['day of year'].apply(lambda x: day_density(x))
    df_temp = pd.merge(df_temp, df_gdp, on=['year', 'country'])
    df_temp = pd.get_dummies(df_temp, columns=['country', 'store', 'product', 'day of week'])
    
    return df_temp
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['day of week'] = df['date'].dt.day_name()
df['month'] = df['date'].dt.month
df['year'] = df['date'].dt.year
```

```python
df['day of year'] = df['date'].apply(lambda x: get_year_day(x))
```

```python
df['holiday density'] = df['day of year'].apply(lambda x: day_density(x))
```

# Analysis

```python
def get_classif(independent, dependent):
    
    dep_array = dependent.array
    
    le = LabelEncoder()
    le.fit(independent)
    temp_encode = le.transform(independent).reshape(-1,1)
    f_val, p_val = f_classif(temp_encode, dep_array)
    
    return f_val, p_val
    
```

```python
def get_mi_reg(independent, dependent):
    
    dep_array = dependent.array
    
    le = LabelEncoder()
    le.fit(independent)
    temp_encode = le.transform(independent).reshape(-1,1)
    p_val = mutual_info_regression(temp_encode, dep_array)
    
    return p_val
```

# Mutual Information

```python
from sklearn.feature_selection import mutual_info_classif, mutual_info_regression
```

```python
df_encode = pd.get_dummies(df, columns=['country', 'store', 'product', 'day of week'])
df_encode.columns
```

```python
X_encode = df_encode.drop(columns=['date', 'num_sold', 'row_id'])
y_encode = df_encode['num_sold']
```

```python
mi_values = mutual_info_regression(X_encode, y_encode, discrete_features=True)
```

```python
mi_dict = {}

for i in range(len(X_encode.columns)):
    mi_dict[X_encode.columns[i]] = mi_values[i]
    
```
